```python
import pandas as pd
```

```python
data = pd.read_excel('cases_data_with_texts_v5.xlsx').drop_duplicates('ID')
data.head()
```

```python
len(data['ID']), len(data['ID'].drop_duplicates())
```

```python
import re
if re.search(r'нан[её]с ей [<\d>]+? удар', 'ей <> ударов'):
    print('slgmlskdm')
```

```python
patterns = [
    r'скончалась',
    r'погибш[аеу][йяю]',
    r'смерт[иь]ю? потерпевшей',
    r'смерт[иь]ю? последней',
    r'мертва',
    r'е[её] труп',
    r'труп женщины'
]

def check_patterns(text):
    global patterns
    
    for pattern in patterns:
        if re.search(pattern, text):
            return True
    return False
```

```python
woman_sample = data.dropna(subset=['text'])[data.dropna(subset=['text'])['text'].apply(check_patterns)].copy()
woman_sample.head()
```

```python
woman_sample.to_excel('women_cases.xlsx')
```

```python
k = 1
for i in range(0, len(woman_sample), 300):
    woman_sample.iloc[i:i+300, :].to_excel(f'women_cases_part_{k}.xlsx')
    k += 1
```
