```python
import pandas as pd
```

```python
import re

patterns = [
    r'скончалась',
    r'погибш[аеу][йяю]',
    r'смерт[иь]ю? потерпевшей',
    r'смерт[иь]ю? последней',
    r'мертва',
    r'е[её] труп',
    r'труп женщины'
]

def check_patterns(text):
    global patterns
    
    for pattern in patterns:
        if re.search(pattern, text):
            return True
    return False
```

```python
def get_fio_MGS(sides):
    pattern = re.search(r'[А-Я][а-я]+ [А-Я]\. ?[А-Я]\.', sides)
    if pattern is not None:
        return pattern[0]
    else:
        pattern = re.search(r'[А-Я][а-я]+ [А-Я]\.', sides)
        if pattern is not None:
            return pattern[0]
        else:
            return None
        
def get_court_name_MGS(cases_list):
    pattern = re.search(r'\([А-яё ]+\)', cases_list)
    if pattern is not None:
        return pattern[0].strip('()')
    else:
        return None
    
def get_article_MGS(sides):
    pattern = re.search(r'\(Ст. \d\d\d, Ч. \d', sides)[0].strip('(')
    return pattern
```

```python
data_MGS = pd.read_excel('MOSGORSUD_texts.xlsx', index_col=0)
data_MGS.columns = [
    'Номер дела (материала)',
    'Стороны',
    'Результат',
    'Судья',
    'Статья УК РФ',
    'Категория дела',
    'Список дел',
    'URL',
    'URL_verdict',
    'ID',
    'text'
]
data_MGS['Источник'] = 'Мосгорсуд'
data_MGS['Субъект РФ'] = 'Москва'
data_MGS['Тип документа'] = 'Приговор'
data_MGS['ФИО'] = data_MGS['Стороны'].apply(get_fio_MGS)
data_MGS['Наименование суда'] = data_MGS['Список дел'].apply(get_court_name_MGS)
data_MGS['Статья УК РФ'] = data_MGS['Стороны'].apply(get_article_MGS)
data_MGS.drop(['Категория дела', 'Список дел'], axis=1, inplace=True)
data_MGS.to_excel('MOSGORSUD_texts_prep.xlsx')
data_MGS.head()
```

```python
articles = {'105': '105', '107': '107', '111_4': '111'}
year = 2018
women_cases = 0
for art in articles:
    data_GAS_art = pd.read_excel(f'{art}_{year}_cases_texts.xlsx', index_col=0)
    data_GAS_art['Источник'] = 'ГАС Правосудие'
    data_MGS_art = data_MGS[data_MGS['Статья УК РФ'].str.contains(articles[art])]
    full_data_art = pd.concat([data_MGS_art, data_GAS_art]).reset_index(drop=True)
    print('Статья', art)
    print(f'ГАС Правосудие: {len(data_GAS_art)}, Мосгорсуд: {len(data_MGS_art)}, всего: {len(full_data_art)}')
    women_cases_art = len(full_data_art[full_data_art['text'].apply(check_patterns)])
    print('Женских дел по статье:', women_cases_art)
    women_cases += women_cases_art
    full_data_art[full_data_art['text'].apply(check_patterns)].to_excel(f'{art}_{year}_women_cases_texts_MERGED.xlsx')
    #full_data_art.to_excel(f'{art}_{year}_cases_texts_MERGED.xlsx')
print('\nЖенских дел всего:', women_cases)
```
